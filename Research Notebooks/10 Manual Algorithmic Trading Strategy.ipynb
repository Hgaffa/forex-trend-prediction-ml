## 10.2 Implementing Technical Indicators

```python
#SSL Indicator Function
def ssl_channel_indicator(df, period = 20):
    
    df['SMA_High'] = originalta.SMA(df.High, timeperiod=period)
    
    df['SMA_Low'] = originalta.SMA(df.Low, timeperiod=period)
    
    hlv = pd.DataFrame(columns=['HLV'], index=df.index)
    
    sslDown = pd.DataFrame(columns=['SSL_Down'], index=df.index)
    
    sslUp = pd.DataFrame(columns=['SSL_Up'], index=df.index)
    
    for i in range(0,len(df.Close)):
        
        if df.Close.iloc[i] > df.SMA_High.iloc[i]:
            
            hlv.iloc[i] = 1
            
        elif df.Close.iloc[i] < df.SMA_Low.iloc[i]:
            
            hlv.iloc[i] = -1
            
        else:
            
            hlv.iloc[i] = hlv.iloc[i-1]
            
    for i in range(0,len(hlv)):
        
        if hlv.HLV.iloc[i] <= 0:
            
            sslDown.iloc[i] = df.SMA_High.iloc[i]
            
        else:
            
            sslDown.iloc[i] = df.SMA_Low.iloc[i]
            
    for i in range(0,len(hlv)):
        
        if hlv.HLV.iloc[i] < 0:
            
            sslUp.iloc[i] = df.SMA_Low.iloc[i]
            
        else:
            
            sslUp.iloc[i] = df.SMA_High.iloc[i]     
    
    return sslDown, sslUp

##################################################################

#ATR Indicator Function
def atr_indicator(df, period = 13):
    
    atr = ta.atr(df.High, df.Low, df.Close, period)

    return atr

##################################################################

#Linear WMA Indicaotr Function
def lwma_indicator(df, period=15):
        
    lwma = pd.DataFrame(columns=['LWMA'], index=df.index)
    
    for i in range(0,len(df)):
        
        if i < (period-1):
            
            continue
        
        else:
            
            sum_prices = 0
    
            for j in range(0,period+1):
            
                price = df.Close.iloc[i-j]
                
                sum_prices += (price * (period-j))
                
            lwma.iloc[i] = sum_prices/(np.sum(range(1,period+1)))
        
    return lwma

##################################################################

#QQE indicators function - similar to RSI indicator (shows trend strength signal)
def qqe_indicator(df, period=10):
    
    qqe = ta.qqe(df.Close, length=period)
    
    qqe.columns = ['QQE','RSI','Long','Short']

    qqe = qqe.reset_index().drop(columns=['Long','Short'])
    
    QQE = qqe['QQE']

    RSI = qqe['RSI']
    
    return QQE.values, RSI.values

##################################################################

#Volatility Index Indicator
def volatility_index_indicator(df, period=10):
    
    volatility = pd.DataFrame(columns=['VOLATILITY'], index=df.index)
    
    for i in reversed(range(0,len(df)-1)):
        
        if df.Close.iloc[i] > df.Open.iloc[i]:
            
            volatility.iloc[i] = abs(df.High.iloc[i] - df.Low.iloc[i+1])
        
        elif df.Close.iloc[i] < df.Open.iloc[i]:
            
            volatility.iloc[i] = abs(df.Low.iloc[i] - df.High.iloc[i+1])
            
        else:
            
            volatility.iloc[i] = volatility.iloc[i+1]
        
    volatility = originalta.SMA(volatility.VOLATILITY, timeperiod=5)
    
    sma_vol = originalta.SMA(volatility, timeperiod=15)
            
    return volatility, sma_vol
```

```python
df = data.copy()

df['Return'] = df.Close.pct_change()

df['SSL_Down'], df['SSL_Up'] = ssl_channel_indicator(df)

df['ATR'] = atr_indicator(df, period=13)

df['LWMA'] = lwma_indicator(df, period=10)

df['QQE'], df['RSI_MA'] = qqe_indicator(df, period=13)

df['VOLA'], df['SMA_VOLA'] = volatility_index_indicator(df, period=13)

df = df.dropna()

df
```

## 10.3 Implementing Trading Strategy

```python
#Function to generate signals from ssl indicator
def ssl_signals(up, down):
    
    return (1 if up > down else 0)


def ssl_trades(df):
    
    predictions = pd.DataFrame(columns=['Predictions'], index=df.index)
    
    for i in range(0, len(df)):
        
        if ssl_signals(df.SSL_Up.iloc[i], df.SSL_Down.iloc[i]) == 1:
            
            predictions.iloc[i] = 1
        
        else:
            
            predictions.iloc[i] = 0
    
    return predictions
###########################################################################

#Function to see if ssl buy signal is new (within 2 observations)
def new_signal(df, index, period):
    
    if df.SSL_Signal.iloc[index] == 1:
    
        for i in range(1,period):

            if df.SSL_Signal.iloc[index-i] == 0:

                return True
    else:
    
        return False

#Function to check if qqe is above 50
def check_qqe(df, index, level):
    
    if df.QQE.iloc[index] > level:
        
        return True
    
    else:
        
        return False
    
#Function to check if moving average is below current daily volatility (high volatility day)
def check_volatility(df, index):
    
    if df.VOLA.iloc[index] > df.SMA_VOLA.iloc[index]:
        
        return True
    
    else:
        
        return False
    
def atr_check(df, index):
    
    if ((df.Close.iloc[index] - df.LWMA.iloc[index]) <= df.ATR.iloc[index]):
        
        return True
    
    else:
        
        return False

    
#Function to check if price is above LWMA baseline
def check_price_baseline(df, index):
    
    if (df.Close.iloc[index] > df.LWMA.iloc[index]):
                
        return True
    
    else:
        
        return False

#Function to long trades from recent ssl signal
def ssl_long_signal(df):
    
    predictions = pd.DataFrame(index=df.index, columns=['Predictions'])
    
    for i in range(0,len(df)):
        
        if i < 3:
            
            predictions.iloc[i] = 0
            
        else:
        
            if check_price_baseline(df, i) and atr_check(df, i) and new_signal(df, i, 3) and check_qqe(df, i, 50) and check_volatility(df, i):
                
                predictions.iloc[i] = 1
                
            else:
                
                predictions.iloc[i] = 0
    
    return predictions

#Function to see if ssl buy signal is well established (within 7 observations)
def established_signal(df, index, period):
    
    for i in range(1,period):
        
        if df.SSL_Signal.iloc[index-i] == 0:
            
            return False
        
        elif df.SSL_Signal.iloc[index-i] == 1:
            
            return True
        
    return False

#Function to signal long from a baseline cross (not new SSL signal)
def baseline_cross_long(df):
    
    predictions = pd.DataFrame(index=df.index, columns=['Predictions'])
    
    for i in range(0,len(df)):
        
        if i < 7:
            
            predictions.iloc[i] = 0
        
        else:
            
            if check_price_baseline(df, i) and established_signal(df, i, 8) and check_qqe(df, i, 50) and check_volatility(df, i):
                
                predictions.iloc[i] = 1
            
            else:
                
                predictions.iloc[i] = 0
        
    return predictions

#Function to check if price has gone over baseline since last long trade
def not_crossed_baseline(df, index, old_preds):
    
    last_trade = -1
    
    for i in reversed(range(0,index+1)):
        
        if old_preds.iloc[i][0] == 1:
            
            last_trade = i
            
            break
    
    if last_trade == -1:
        
        return False
    
    else:
        
        for j in range(last_trade, index+1):
            
            if check_price_baseline(df, j):
                
                return False
            
        return True
            

#Function to signal continuation trade on a new SSL signal
def continue_long_ssl_signal(df, old_predictions):
    
    predictions = pd.DataFrame(index=df.index, columns=['Predictions'])
    
    for i in range(0,len(df)):
        
        if check_price_baseline(df, i) and not_crossed_baseline(df, i, old_predictions) and (df.SSL_Signal.iloc[i] == 1):
            
            predictions.iloc[i] = 1
            
        else:
            
            predictions.iloc[i] = 0
            
    return predictions
                
###########################################################################

#Report Accuracy function
def report_accuracy(predictions, targets):
    
    true_pos = 0
    
    false_pos = 0
    
    true_neg = 0
    
    false_neg = 0

    for i in range(0, len(targets)):
        
        if((predictions.iloc[i].values == 1) and (targets.iloc[i].values == 1)):
            
            true_pos += 1
            
        elif((predictions.iloc[i].values == 1) and (targets.iloc[i].values == 0)):
            
            false_pos +=1
            
        elif((predictions.iloc[i].values == 0)and (targets.iloc[i].values == 0)):
            
            true_neg += 1
            
        else:
            
            false_neg += 1
 
    print("True Positives: ", true_pos)
    
    print("False Positives: ", false_pos)
    
    print("True Negatives: ", true_neg)
    
    print("False Negatives: ", false_neg)
    
    print("True Positive %: ", true_pos/(false_pos+true_pos)*100)
    
def financial_report(predictions, returns):
    
    cum_returns = pd.DataFrame(index=predictions.index, columns=['Returns'])
    
    total = 1000
    
    for i in range(0,len(predictions)-1):
    
        total += (predictions.iloc[i]*returns.iloc[i+1])*total

        cum_returns.iloc[i] = (total.values[0] - 1000)/1000
        
    cum_returns = cum_returns[cum_returns['Returns'] != 0]
    
    r = cum_returns.diff()
    
    r = r.replace([np.inf, -np.inf], np.nan)
    
    r = r.dropna()
    
    sr = r.mean()/r.std() * np.sqrt(252)
    
    return sr, total
```
