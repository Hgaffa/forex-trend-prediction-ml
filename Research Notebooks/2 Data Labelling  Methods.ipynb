## 2.2 Applying Triple Barrier Method

Here I apply the triple barrier method to the processed dataset to produce a set of labels for the training data. The idea behind the triple barrier method is to incorporate the intermediate price values between the horizon being used to calculate the return, this allows for the identification of price trends more accurately, without completely ignoring the intricate market movements

```python
#ATR Indicator Function to calculate daily volatility to be able calculate the two horizontal barriers (for SL and TP thresholds)  
def atr_indicator(df, period = 13):
    
    atr = ta.atr(df.High, df.Low, df.Close, period)

    return atr

dailyvol = atr_indicator(df)

daily_vol.dropna(inplace=True)

daily_vol.head()
```

```python
#define the period at which to set the vertical barrier at
hold = 10

#period to calculate returns for FHM
period = 1

prices = prices[daily_vol.index]

#triple barrier function
def triple_barrier(dv, sl_mult, tp_mult, time_to_hold):
    
    #create dataframe for barriers
    barriers = pd.DataFrame(columns=['rate','vertical_barrier','top_barrier','bottom_barrier'],index=dv.index)
    
    n = 0
    
    for day, vol in dv.iteritems():
        
        n += 1    
        
        #Get vertical barrier
        if (n + time_to_hold < len(dv.index) and time_to_hold >= 0):
            
            vertical_barrier = dv.index[n + time_to_hold]
            
        else:
            
            vertical_barrier = np.nan
            
        #Set top barrier for TP
        if tp_mult > 0:
            
            top_barrier = prices.loc[day] + prices.loc[day] * tp_mult * vol
            
        #Set bottom barrier for SL
        if sl_mult > 0:
            
            bottom_barrier = prices.loc[day] - prices.loc[day] * sl_mult * vol
            
        barriers.loc[day, ['rate','vertical_barrier','top_barrier','bottom_barrier']] = prices.loc[day], vertical_barrier, top_barrier, bottom_barrier   
        
    return barriers

barriers = triple_barrier(daily_vol, 2, 2, 10)

barriers
```

It can be noted that the vertical barrier is held as a data, this is so that price between the current observation date and the date of the final vertical barrier can be checked. If neither horizontal barriers are hit between this data range then it is assume to be a ranging period.

```python
barriers.info()
```

From the above information it can be seen that only the vertical barrier has 11 NaN values which corresponds to the 10 day period used when calculating it. The idea now is to find which barrier is hit first and apply the corresponding label to the data observation at that index. If the end barrier is hit without hitting either horizontal barriers then the Fixed Horizon Method is employed due to the problem being a binary classification case. This is the easiest way to determine binary labels for non-trending regions and produces a hybrid approach to labeling, taking the benefits of both techniques.

```python
#Find returns of AUD/USD price series for Fixed Horizon Labelling
returns = barriers.rate.pct_change()
returns
```

## 2.3 Generating Labels

```python
#function to generate labels from ascertaining which barrier is hit first
def generate_labels(barriers):
    
    barriers['label'] = None
    
    non_trends = 0
    ups = 0
    downs = 0
    
    for i in range(len(barriers.index)):
        
        current = barriers.index[i]
        
        vert = barriers.vertical_barrier[i]
        
        if pd.notna(vert):
            
            #assign initial and final prices
            current_rate = barriers.rate[current]
            
            vert_rate = barriers.rate[vert]
            
            #assign top and bottom barriers
            tb = barriers.top_barrier[i]
            
            bb = barriers.bottom_barrier[i]
            
            #calculate SL and TP conditions
            take_profit = (barriers.rate[current : vert] >= tb)
            
            stop_loss = (barriers.rate[current : vert] <= bb)
            
            tp = False
            sl = False
            
            if take_profit.any():
                tp = True
            elif stop_loss.any():
                sl = True
            
            #generate labels
            if tp:
                
                barriers['label'].iloc[i] = 1
                ups += 1
                
            elif sl:
                
                barriers['label'].iloc[i] = -1
                downs += 1
            
            else:
                
                #Fixed Horizon Method
                if (returns.iloc[i+1]) > (returns.iloc[i]):
                    barriers['label'].iloc[i] = 1
                else:
                    barriers['label'].iloc[i] = -1
    
                #count number of potentially non trending regions
                non_trends += 1
    
    return non_trends, ups, downs, barriers;

non_trends, ups, downs, barriers = generate_labels(barriers)
barriers
```

```python
#count how many TP and SL barriers were hit throughout the dataset
print("Triple Barrier Method Labels for AUD/USD dataset")
barriers.label.value_counts()
```

```python
#calculate labels from pure Fixed Horizon Method with period = 1
print("Fixed Horizon Method Labels for AUD/USD dataset")
fhm_labels = pd.DataFrame(returns.shift(-1) - returns).applymap(lambda x: 1 if x > 0 else - 1)
fhm_labels.value_counts()
```

```python
print("Number of non-trending regions within AUD/USD dataset:", non_trends)
print("Number of take profits:", ups)
print("Number of stop losses:", downs)
```

It can be observed that the number of non-trending regions within the dataset is 676/2911. This is a high proportion of non-trending regions that are handled by the Fixed Horizon Method.

```python
#Plotting employed Triple Barrier Labelling Method for AUD/USD dataset
def plot_TBM(barriers, sl_mult=0, tp_mult=0):
    fig,ax = plt.subplots(figsize=(20,10))

    ax.set_title("AUD/USD Triple Barrier Method")
    ax.set_xlabel("Date")
    ax.set_ylabel("Close Price")

    ax.plot(barriers.rate[100: 300])

    current_val = barriers.index[180]
    future_period = barriers.vertical_barrier[210]
    top_b = barriers.top_barrier[180] + barriers.rate[180] * tp_mult
    bot_b = barriers.bottom_barrier[180] + barriers.rate[180] * -sl_mult

    #plot top barrier
    ax.plot([current_val, future_period], [top_b, top_b], linestyle='--', c='green', label='top_barrier')

    #plot bottom barrier
    ax.plot([current_val, future_period], [bot_b, bot_b], linestyle='--', c='red', label='bottom_barrier')

    #plot central line of action
    ax.plot([current_val, future_period], [(bot_b + top_b)*0.5, (bot_b + top_b)*0.5], c='black', linestyle='--')

    #plot line for current value
    ax.plot([current_val, current_val], [bot_b, top_b], linestyle='--', c='black')

    #plot vertical barrier after holding period
    ax.plot([future_period, future_period], [bot_b, top_b], linestyle='--', c='blue', label='vertical_barrier')
    
    ax.legend()
    
plot_TBM(barriers)
```

The above plot demonstrates the triple barrier method in which the derived label for the marked observation would be an downtrend as the bottom barrier is hit first. The height of each barrier can change depending on the multiplier supplied, making them model tunable in terms of sensitivity to trends. Below it can be observed how the barriers change by changing the multiplier to 0.02 of the current close price.

```python
plot_TBM(barriers, 0.02, 0.02)
```

It can be observed how the generated label changes to an uptrend by simply changing the multiplier of the horizontal boundaries. Hence, this approach is highly tunable to the problem domain.

## 2.5 Varying Period Size for FHM & Holding Time for TBM

The fundamental hyperparameter for both labeling methods is the period. This value determines how many past or future values are used in the calculation of labels. For example, in the previous case, next day trends were generated as labels for the FHM by using a period of 1 on 1 percentage returns. By changing the values for the period and the number of steps predicted in the future, different labelling results can be observed.

```python
#Doubling holding time for TBM
def labelling_test(tbm_holding_period, fhm_period, fhm_forward):

    barriers = triple_barrier(daily_vol, 2, 2, tbm_holding_period)
    
    non_trends, ups, downs, barriers = generate_labels(barriers)
    
    returns = barriers.rate.pct_change(fhm_period)
    
    fhm_labels = pd.DataFrame(returns.shift(-fhm_forward) - returns).applymap(lambda x: 1 if x > 0 else - 1)
    
    print("Number of non-trending regions within AUD/USD dataset:", non_trends)
    print("Number of take profits:", ups)
    print("Number of stop losses:", downs)
    
    print("\nTriple Barrier Method Labels for AUD/USD dataset with holding period:", tbm_holding_period)
    print(barriers.label.value_counts())
    
    print("\nFixed Horizon Method Labels for AUD/USD dataset with back/forward periods:", fhm_period, fhm_forward)
    print(fhm_labels.value_counts())

labelling_test(20, 2, 2)
```

By doubling the holding period for the TBM, it can be observed that the number of non-trending regions dramatically decreased to 257. However, the overall outlook of the market has completely shifted to a positive one with 1581 uptrends and 1320 downtrends. This is the complete opposite result to the FHM and the general trend of the AUD/USD dataset. Moreover, care must be taken when using the TBM and tuning of the hyperparameters must be considered.

## 2.6 Comparing Accuracy of Labels on KNN Classifier

Here, the labels generated by the Fixed Horizon and Triple Barrier Method are evaluated in terms of classification accuracy. This evaluation determines how well the methods can capture trends relative to price. Note that the only features used for prediction are the OHLC pricing data with labels being generated such that next day trends are predicted for FHM method and a holding period of 10 is used for TBM. 

```python
data = df.copy()

#get labels
def get_labels(data, method):
    #get barriers and returns
    barriers = triple_barrier(daily_vol, 2, 2, 10)

    non_trends, ups, downs, barriers = generate_labels(barriers)

    returns = barriers.rate.pct_change(1)

    fhm_labels = pd.DataFrame(returns.shift(-1)).applymap(lambda x: 1 if x > 0 else - 1)

    if method == "TBM":
        labels = barriers.label
    elif method == "FHM":
        labels = fhm_labels
    else:
        print("Not real method")
        return
    
    labels = labels[labels.notna()]

    #make indexing for price data the same as the generated labels
    joined = pd.concat([data, labels], axis=1)

    joined.dropna(inplace=True)
    
    joined.columns = ['Open','High','Low','Close','Volume','Label']
    
    return joined

def knn_model(data, labels, method):
    #generate training data and split
    X = data.copy().drop(columns=['Label'])
    y = labels.values.astype(int)
    
    split = int(len(X)*0.7)

    X_train = X[0:split]
    X_test = X[split+1:]
    
    y_train = y[0:split]
    y_test = y[split+1:]

    knn = KNeighborsClassifier()
    knn.fit(X_train,y_train)
    predicted = knn.predict(X_test)

    score = cross_val_score(knn,X_train,y_train)
    
    print("The cross validation score of the " + method + " is", knn.score(X_test,y_test))
    
    #Plot confusion matrices
    titles_options = [("Confusion matrix, without normalization", None)]

    disp = plot_confusion_matrix(knn, X_test, y_test,                           
                                 display_labels=[-1,1],
                                 cmap=plt.cm.Blues)

    disp.ax_.set_title("Confusion Matrix")
    
    return score
```

```python
#Fixed horizon method results
df2 = get_labels(data,"FHM")
knn_model(df2, df2.Label, "Fixed Horizon Method")
```

```python
#Fixed horizon method results
df3 = get_labels(data,"TBM")
knn_model(df3, df3.Label, "Triple Barrier Method")
```

From the basic results from the KNN classifier above, the Fixed Horizon Method has shown to achieve a higher average classification accuracy used standard next day trends. This contradicts the initial idea of using the TBM as a more accuracy labelling technique. However, it is important to note that this is simply a classification evaluation, the financial viability of such a model is not shown and could be where the TBM excels.
